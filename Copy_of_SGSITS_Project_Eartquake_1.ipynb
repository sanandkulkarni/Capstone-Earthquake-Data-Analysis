```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('database.csv')
data.head(20)
```

```python
data.isnull().sum()
```

```python
data = data.drop(['Depth Seismic Stations','Magnitude Seismic Stations','Azimuthal Gap','Horizontal Error','Horizontal Distance','Depth Error'],axis=1)
```

```python
data
```

```python
data['Date']=pd.to_datetime(data["Date"],errors='coerce')
```

```python
data
```

```python
data['Year'] = data['Date'].dt.year
```

```python
data['Month']= data['Date'].dt.month
data
```

```python
print(data['Date'].isnull().sum())
```

```python
data.dropna(subset= ["Date"],inplace=True)
```

```python
data.isnull().sum()
```

```python
data = data.drop(['Magnitude Type','Magnitude Error'],axis=1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['Root Mean Square'].fillna(data['Root Mean Square'].median())
```

```python
data.isnull().sum()
```

```python
data['Magnitude'].max()
```

```python
data['Depth'].max()
```

```python
data['Custom_Label'] = data.apply(
    lambda row: f"{row['Date'].strftime('%Y-%m-%d')}\n({row['Latitude']}, {row['Longitude']})", axis=1
)
```

```python
Lowest_magnitudes = data.nsmallest(5,'Magnitude')

plt.figure(figsize=(10, 5))
plt.bar(Lowest_magnitudes['Custom_Label'], Lowest_magnitudes['Magnitude'], color='salmon')
plt.title('5 Highest Magnitude Earthquakes')
plt.xlabel('Date and Location')
plt.ylabel('Magnitude')
plt.xticks(rotation=0, fontsize=8)  # no rotation needed now
plt.tight_layout()
plt.show()
```

```python
Lowest_magnitudes = data.nsmallest(5,'Magnitude')
Lowest_magnitudes
```

```python
Highest_magnitudes = data.nlargest(5,'Magnitude')
Highest_magnitudes
```

```python
plt.figure(figsize=(10,5))
plt.bar(Highest_magnitudes['Custom_Label'],Highest_magnitudes['Magnitude'],color='skyblue')
plt.title('5 Highest Magnitude Earthquake')
plt.xlabel('Location and Date')
plt.ylabel('Magnitude')
plt.show()
```

```python
earthquakes_per_year=data['Year'].value_counts().sort_index()
earthquakes_per_year
```

```python
plt.figure(figsize=(12,5))
plt.bar(earthquakes_per_year.index.astype(str),earthquakes_per_year.values,color='steelblue')
plt.title('Number of Earthquakes per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45, fontsize=8)
plt.tight_layout()
plt.show()
```
